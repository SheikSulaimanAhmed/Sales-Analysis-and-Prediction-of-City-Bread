# SALES PREDICTION AND ANALYSIS OF CITY BREAD

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.metrics import accuracy_score, confusion_matrix
%matplotlib inline
```

```python
data = pd.read_csv('Book1.csv')
print("Data imported successfully!!")
```

```python
data.head(20)
```

## White Loaf

```python
df = pd.DataFrame(data)
cols = [1,3]
df = df[df.columns[cols]]
```

```python
df.head(10)
```

```python
data.plot(x='MONTH', y='WHITE LOAF', style='-', color='blue')
plt.title('Month vs Sales of White Loaf') 
plt.xlabel('Month')  
plt.ylabel('Number of Products sold') 
plt.show()
```

```python
X = df.iloc[:, :-1].values 
y = df.iloc[:, 1].values  
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print("Training complete.")
```

```python
#Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data\n"
plt.scatter(X, y, color='red')
plt.plot(X, line, color='black');
plt.title("Comparison")
plt.show()
```

```python
print(X_test) 
y_pred = regressor.predict(X_test) # Predicting the values
```

```python
# Comparing Actual vs Predicted
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 
```

```python
score_pred=np.array([12])
score_pred=score_pred.reshape(-1,1)
predict=regressor.predict(score_pred)
print("Month={}".format(12))
print("Predicted Score for White Loaf={}".format(predict[0]))
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```

```python
mape = np.mean(np.abs((y_test - y_pred)/y_test))*100
mape
```
